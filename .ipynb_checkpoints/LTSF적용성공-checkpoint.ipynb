```python
train_df = pd.read_csv('./train.csv')
test_df = pd.read_csv('./test.csv')
sample_submission = pd.read_csv('./sample_submission.csv')
# 일조, 일사 열 제거
train_df = train_df.drop(['일조(hr)','일사(MJ/m2)'], axis=1)
train_df.head()

# 강수량 결측치 0.0으로 채우기
train_df['강수량(mm)'].fillna(0.0, inplace=True)

# 풍속, 습도 결측치 평균으로 채우고 반올림하기
train_df['풍속(m/s)'].fillna(round(train_df['풍속(m/s)'].mean(),2), inplace=True)
train_df['습도(%)'].fillna(round(train_df['습도(%)'].mean(),2), inplace=True)
# 데이터 변환
df = train_df.copy()
df.rename(columns={'num_date_time': 'date_time'}, inplace=True)

df['date_time'] = df['date_time'].str.replace('.*_', '', regex=True)

df['date_time'] = pd.to_datetime(df['date_time'], format='%Y%m%d %H')
```

```python
pred_list= []
for building_number in range(1, 101):
    # '건물번호'가 현재 building_number인 행 추출
    filtered_df = df[df['건물번호'] == building_number]
    
    # '건물번호'열 삭제
    filtered_df = filtered_df.drop(columns=['건물번호'])
    
    # raw에 데이터 저장
    raw = filtered_df
    
    # split_data 함수로 날짜와 데이터 분리
    date, data = split_data(raw, '전력소비량(kWh)')
    
    # 항상 확정된 window_size, forecast_size
    window_size, forecast_size = 24, 186
    
    # build_dataLoader 함수로 데이터 로더 구축
    dataloader = build_dataLoader(data, window_size=window_size, forecast_size=forecast_size, batch_size=4)
    
    # trainer 함수로 예측값 획득
    pred = trainer(data, dataloader, window_size=window_size, forecast_size=forecast_size).implement()
    # pred 값을 16800개로 확장하여 쌓기
    pred_list.append(pred)
    # 결과를 test_df에 저장
 #   test_df = pd.concat([test_df, pd.DataFrame({'Date': date, '전력소비량(kWh)': pred})], ignore_index=True)
```
